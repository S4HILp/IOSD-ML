```python
import pandas as pd
import numpy as np

df = pd.read_csv('heartdisease.csv')
df.drop(columns='education',inplace=True)
df
```

```python
from sklearn.svm import SVC

model = SVC(C=10)

data = df[['age','prevalentHyp','sysBP','diaBP']] 
y = df['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(data,y,test_size=0.4,random_state=0)

model.fit(x_train,y_train)

pred = model.predict(x_test)

model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score

acc = cross_val_score(model,data,y,cv=10)
acc.mean()
#--> AVERAGE ACCURACY
```

```python
from sklearn import metrics

metrics.confusion_matrix(y_test,pred,labels=(1,0))

#--> SINCE THE VALUES IN TP AND FP ARE ZERO WE GET A ZERO DIVISION WARNING BELOW
```

```python
print(metrics.classification_report(y_test,pred))
```
