```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

df = pd.read_csv('car_price.csv')
df['fuel'].replace({'Petrol':1.,'Diesel':2.,'CNG':3.,'LPG':4.},inplace=True)
df['transmission'].replace({'Manual':1.,'Automatic':2,},inplace=True)
df['owner'].replace({'First Owner':1.,'Second Owner':2.,'Third Owner':3.,'Fourth & Above Owner':4.,'Test Drive Car':5.},inplace=True)
df
```

```python
company = df['name'].apply(lambda x: x.split(' ')[0])
df.insert(2,'company',company)
```

```python
df.drop(columns='name',inplace=True)
```

```python
# ALL MODELS TOGETHER

model_0 = LinearRegression()

x0_train,x0_test,y0_train,y0_test = train_test_split(df[['year','km_driven','transmission','owner']],df['selling_price'],test_size=0.3,random_state=0)

model_0.fit(x0_train,y0_train)

predicted0 = model_0.predict(x0_test)
predicted0
```

```python
from sklearn.metrics import r2_score

r2_score(y0_test,predicted0) #--> ACCUARCY OF THE MODEL
```

```python
#FOR A SINGLE MODEL

car_model = df.groupby(['company']
data = car_model[['year','km_driven','owner']].get_group('Maruti')
year = car_model['year'].get_group('Maruti')
kms = car_model['km_driven'].get_group('Maruti')
own = car_model['owner'].get_group('Maruti')
price = car_model['selling_price'].get_group('Maruti')
```

```python
model_1 = LinearRegression()

x1_train,x1_test,y1_train,y1_test = train_test_split(data,price,test_size=0.66,random_state=0)

model_1.fit(x1_train,y1_train)

predicted1 = model_1.predict(x1_test)
predicted1
```

```python
from sklearn import metrics

r2_score(y1_test,predicted1)

#--> ACCURACY OF THE MODEL
```
